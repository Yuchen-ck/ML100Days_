```python
import pandas as pd
score_df = pd.DataFrame([[1,50,80,70,'boy',1], 
              [2,60,45,50,'boy',2],
              [3,98,43,55,'boy',1],
              [4,70,69,89,'boy',2],
              [5,56,79,60,'girl',1],
              [6,60,68,55,'girl',2],
              [7,45,70,77,'girl',1],
              [8,55,77,76,'girl',2],
              [9,25,57,60,'girl',1],
              [10,88,40,43,'girl',3],
              [11,25,60,45,'boy',3],
              [12,80,60,23,'boy',3],
              [13,20,90,66,'girl',3],
              [14,50,50,50,'girl',3],
              [15,89,67,77,'girl',3]],columns=['student_id','math_score','english_score','chinese_score','sex','class'])
score_df
```

```python
boy_score_df = score_df.loc[score_df.sex=='boy']
girl_score_df = score_df.loc[score_df.sex=='girl']
print(boy_score_df.mean())
print()
print(girl_score_df.mean())
```

```python
score_df.groupby('sex').mean()
#groupby : 從不同的sex中計算相應欄位數值
```

```python
#1.找出全年級各科成績最高分與最低分?
print("math")
print(score_df.math_score.max())
print(score_df.math_score.min())
print()

print("english")
print(score_df.english_score.max())
print(score_df.english_score.min())
print()

print("chinese")
print(score_df.chinese_score.max())
print(score_df.chinese_score.min())
```

```python
#2.找出數學班平均最高的班級?
score_df.groupby('class').math_score.max()
```

```python
score_df.groupby('class').mean()
#答案: 2班
```

```python
score_df.groupby(['class','sex']).mean()
```

```python
score_df.groupby(['class','sex']).agg(['max','min'])
```

```python
#3.分析全校女生與男生國文平均差幾分?
score_df.groupby(['sex']).chinese_score.mean()
```
